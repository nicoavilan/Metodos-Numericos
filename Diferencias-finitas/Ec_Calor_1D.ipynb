# Dirchlet

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fil = 500
col = 20

M = np.zeros((fil,col))

L = 2
T = 3
k = 0.1

ht = T/(fil-1)
hx = L/(col-1)
print("ht=",ht)
print("hx=",hx)

x = np.linspace(0,2,col)
M[0,:] = np.sin(np.pi*x/L).round(3)

plt.plot(x,M[0,:])
plt.show()
for n in range(fil-1):
  for m in range(1,col-1):
    M[n+1,m] = M[n,m] + (k*ht/hx**2)*(M[n,m+1] -2*M[n,m] + M[n,m-1])
  plt.plot(x,M[n,:])
plt.show()

plt.plot(x,M[-1,:])
plt.show()
```

## Animación

```python
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
```

```python
fig, ax = plt.subplots()

ax.set_xlim(( 0, L))
ax.set_ylim((0, 1))

line, = ax.plot([], [], lw=3, color="red")
ax.set_title("Solución f(t,x)");
```

```python
def init():
    line.set_data([], [])
    return (line,)

def animate(i):
    y = M[i,:]
    line.set_data(x, y)
    return (line,)

anim = FuncAnimation(fig, animate, init_func=init,
                               frames=fil, interval=100, blit=True)
HTML(anim.to_html5_video())
```

```python
fil = 500
col = 20

M = np.zeros((fil,col))
#print(M)

L = 2
T = 5
k = 0.1

ht = T/(fil-1)
hx = L/(col-1)
print("ht=",ht)
print("hx=",hx)

x = np.linspace(0,2,col)
M[0,10] = 1
#M[0,:] = np.sin(np.pi*x/L).round(3)
#print(M)

plt.plot(x,M[0,:])
plt.show()
for n in range(fil-1):
  for m in range(1,col-1):
    M[n+1,m] = M[n,m] + (k*ht/hx**2)*(M[n,m+1] -2*M[n,m] + M[n,m-1])
  M[n+1,0] = M[n+1,1]
  M[n+1,-1] = M[n+1,-2]
  plt.plot(x,M[n,:])
plt.show()

plt.plot(x,M[-100,:])
plt.show()
```
